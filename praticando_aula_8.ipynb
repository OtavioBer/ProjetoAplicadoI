```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import metrics
from sklearn import feature_selection

# Carregar conjunto de dados (exemplo: dados_banco)
dados = {'idade': [30, 25, 40, 35, 45,],
         'saldo_conta': [5000, 3000, 10000, 8000, 6000],
         'compras_mes': [8, 10, 5, 7, 6],
         'inadimplente': [True, True, False, False, True]}

df = pd.DataFrame(dados)

# Calcular a importância dos recursos usando critérios de informação mútua
mutual = feature_selection.mutual_info_classif(df.drop(columns='inadimplente'), df.inadimplente)

# Criar gráfico de barras horizontais
f = plt.figure(figsize=(8, 5))
plt.barh(df.columns[:-1], mutual, color='g', alpha=0.6)
plt.title('Importância do Recurso - Critérios de Informação Mútua', fontsize=14, weight='bold')
plt.yticks(fontsize=12, rotation=60)
plt.xlabel('Informações mútuas relacionadas a `inadimplente`', fontsize=12, weight='bold')
plt.ylabel('Atributo', fontsize=12, weight='bold')
plt.show()
```
